```python
# 16. Set Operations:
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
lst1= [x for x in range(1,21,2)]
lst2=[x for x in range(2,21)]
union_lst=[]
def union(lst1,lst2):
    for x in lst1:
        union_lst.append(x)
    for y in lst2:
        if y not in lst1:
            union_lst.append(y)
        else:
            continue
    return sorted(union_lst)
intersection_lst= []
def intersection(lst1,lst2):
    for x in lst2:
        if x in lst1:
            intersection_lst.append(x)
    return intersection_lst
symmetric_diff=[]
def symmetric_difference(union_lst, intersection_lst):
    for y in union_lst:
        if y in intersection_lst:
            continue
        else:
            symmetric_diff.append(y)
    return symmetric_diff
            
print(union(lst1,lst2))
print(intersection(lst1,lst2))
print(symmetric_difference(union_lst,intersection_lst))

venn2( subsets= (set(lst1), set(lst2)) , set_labels = ('lst1','lst2'), set_colors=("yellow","red"), alpha=0.7)
plt.show()

```

```python
# 22. Normalization:
def normalise(lst):
    x= min(lst)
    y= max(lst)
    lst_normalise=[]
    for k in lst:
        lst_normalise.append((k-x)/(y-x))
    return lst_normalise
print(normalise([1,2,3,5]))
lst= [1,2,3,5]
import matplotlib.pyplot as plt
import numpy as np
xpts= np.array([1,2,3,5])
ypts=np.array(normalise(lst))
plt.plot(xpts,ypts)
plt.grid(True)
plt.show()

    

```

```python
# 17. File Operations:
def write_numbers_to_file(filename, numbers):
    f= open('filename', 'w')
    for number in numbers:
        f.write(f"{number}\n")

def read_numbers_and_compute_sum(filename):
    f= open('filename', 'r')
    numbers = [float(line.strip()) for line in f]
    return sum(numbers)
filename = "numbers.txt"
numbers = [10, 20, 30, 40, 50]  
write_numbers_to_file(filename, numbers)
total_sum = read_numbers_and_compute_sum(filename)
print(f"The sum of the numbers in {filename} is: {total_sum}")
```

```python
# 18. Word Count and Frequency:
def writefile(filename, strng):
    f= open(filename, "w")
    f.write(strng)
    f.close()
def countwords(filename, strng):
    dictt={}
    f= open(filename, "r")
    k=len(strng.split())
    z=list(set(strng.split()))
    for y in z:
        dictt[y]= (strng.split()).count(y)
    return max(dictt, key=dictt.get)
import matplotlib.pyplot as plt
import numpy as np
def plotbar(filename, strng):
    dictt={}
    f= open(filename, "r")
    k=len(strng.split())
    z=list(set(strng.split()))
    for y in z:
        dictt[y]= (strng.split()).count(y)
    x= np.array(list(dictt.keys()))
    y= np.array(list(dictt.values()))
    plt.bar(x,y)
    plt.show()
    
    

writefile("neel.txt", "the quick quick brown fox fox jumped over the lazy dog")
print(countwords("neel.txt", "the quick quick brown fox fox jumped over the lazy dog" ))
print(plotbar("neel.txt", "the quick quick brown fox fox jumped over the lazy dog"))
```
